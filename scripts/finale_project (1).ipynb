# Imputation des valeurs manquantes

```python
# ----------------------------
# 1️⃣ Préparation
# ----------------------------

# Trier par pays et année
df = df_filtered.sort_values(['country_name', 'year'])

# Colonnes numériques à imputer
cols_to_impute = df.columns[3:]  # à adapter selon ton dataset

# ----------------------------
# 2️⃣ Interpolation temporelle par pays
# ----------------------------

for col in cols_to_impute:
    df[col] = df.groupby('country_name')[col].transform(
        lambda x: x.interpolate(method='linear', limit_direction='both')
    )

# ----------------------------
# 3️⃣ Imputation multivariée itérative simplifiée
# ----------------------------

# Nombre d'itérations (similaire à IterativeImputer)
n_iter = 5

for i in range(n_iter):
    # Recalculate correlation matrix in each iteration, as imputed values change
    corr = df[cols_to_impute].corr()

    for target in cols_to_impute:
        # Skip if target column has no NaN values left
        if df[target].isna().sum() == 0:
            continue


        if corr[target].drop(target).abs().max() == 0 or pd.isna(corr[target].drop(target).abs().max()):
             continue

        best_predictor = corr[target].drop(target).abs().idxmax()

        # Prepare data for polyfit, only using rows where target is NOT NaN
        x_fit = df.loc[df[target].notna(), best_predictor]
        y_fit = df.loc[df[target].notna(), target]


        # polyfit(x, y, 1) requires at least 2 unique x values
        if len(x_fit.dropna().unique()) < 2:
            continue

        mask_impute = df[target].isna() & df[best_predictor].notna()

        # If there are no values to impute in this iteration with this method, skip
        if mask_impute.sum() == 0:
            continue

        try:
            # Perform linear regression: y = a*x + b
            a, b = np.polyfit(x_fit, y_fit, 1)

            # Impute missing values
            df.loc[mask_impute, target] = a * df.loc[mask_impute, best_predictor] + b
        except np.linalg.LinAlgError:
            continue

# ----------------------------
# 4️⃣ Sécurité : valeurs restantes
# ----------------------------

for col in cols_to_impute:
    df[col] = df[col].fillna(df[col].median())

# ----------------------------
# 5️⃣ Vérification finale
# ----------------------------

print("Valeurs manquantes par colonne après imputation :")
print(df.isnull().sum())
```

## normalisation

```python
cols_no_norm=['country_name', 'country_code', 'year']
```

```python
# Select only numeric columns for normalization, excluding the identifier columns
numeric_cols = df.select_dtypes(include=np.number).columns
cols_to_norm = [col for col in numeric_cols if col not in cols_no_norm]
cols_to_norm
```

```python
# Fonction Min-Max
def normalize(df, columns):
    result = df.copy()
    for feature_name in columns:
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result

# Application
df_norm = normalize(df, cols_to_norm)
```

# Definition des score

## Scale Score (Taille du Marché) - Poids : 35%

```python
# 2. Groupe "Effectifs" (Volumes absolus)
effectifs_cols = [
    'population,_ages_15-24,_total',
    'enrolment_in_secondary_education,_both_sexes_(number)',
    'enrolment_in_tertiary_education,_all_programmes,_both_sexes_(number)'
]
df_norm['avg_volumes'] = df_norm[effectifs_cols].mean(axis=1)

# 3. Groupe "Ratios" (Intensité/Maturité)
ratios_cols = [
    'gross_enrolment_ratio,_secondary,_both_sexes_(percent)',
    'gross_enrolment_ratio,_tertiary,_both_sexes_(percent)'
]
df_norm['avg_ratios'] = df_norm[ratios_cols].mean(axis=1)

# 4. Calcul final du Scale Score (sur 100)
df_norm['scale_score'] = (
    (df_norm['TAM_connected_youth'] * 0.50) +
    (df_norm['avg_volumes'] * 0.30) +
    (df_norm['avg_ratios'] * 0.20)
) * 100

df_norm[['country_name','scale_score']].sort_values(ascending=False, by='scale_score').head()
```

# Market Attractiveness Score (MAS)


```python
# 1. Définition des poids (Weights)
w_scale = 0.35
w_momentum = 0.25
w_ability = 0.25
w_digital = 0.15

# 2. Calcul du MAS
# On s'assure que chaque score de pilier est bien sur 100
df_norm['MAS'] = (
    (df_norm['scale_score'] * w_scale) +
    (df_norm['Momentum_Score'] * w_momentum) +
    (df_norm['ability_score'] * w_ability) +
    (df_norm['digital_readiness_score'] * w_digital)
)
```

```python
df_ranked=df_norm.sort_values('MAS', ascending=False)
df_ranked[['country_name','year','MAS']].head(10)
```

```python
# 3. Segmentation Stratégique (Tiering)
conditions = [
    (df_norm['MAS'] >= 70),
    (df_norm['MAS'] >= 45) & (df_norm['MAS'] < 70),
    (df_norm['MAS'] < 45)
]
choices = ['Tier 1: Go-to-Market', 'Tier 2: Build & Partnerships', 'Tier 3: Long-term/Niche']

df_norm['market_priority'] = np.select(conditions, choices, default='Unclassified')
```

```python
import plotly.graph_objects as go

def plot_country_radar(country_name, df):
    data = df[(df['country_name'] == country_name) & (df['year'] == last_year)]
    categories = ['Scale', 'Momentum', 'Ability to Pay', 'Digital']

    fig = go.Figure()
    fig.add_trace(go.Scatterpolar(
          r=[data['scale_score'].values[0], data['Momentum_Score'].values[0],
             data['ability_score'].values[0], data['digital_readiness_score'].values[0]],
          theta=categories,
          fill='toself',
          name=country_name
    ))

    fig.update_layout(polar=dict(radialaxis=dict(visible=True, range=[0, 100])),
                      showlegend=True, title=f"Profil de Performance : {country_name}")
    fig.show()

import matplotlib.pyplot as plt
import numpy as np

pays = [
    "South Africa",
    "Nigeria",
    "Mauritius",
    "Botswana",
    "Ghana",
    "Cabo Verde",
    "Ethiopia",
    "Seychelles",
    "Zimbabwe",
    "Lesotho"
]

plt.figure(figsize=(12, 8))
for country in pays:
    plot_country_radar(country, df_final)





```
